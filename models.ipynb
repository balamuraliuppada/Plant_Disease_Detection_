```python
import tensorflow as tf

# Load old rice model
old_model = tf.keras.models.load_model(
    "models/rice_model_fixed.keras",
    compile=False,
    safe_mode=False
)

# Rebuild model manually from layers (unwrap)
inputs = tf.keras.Input(shape=(224,224,3))

x = inputs

for layer in old_model.layers:
    # Skip old input layer
    if isinstance(layer, tf.keras.layers.InputLayer):
        continue

    x = layer(x)

new_model = tf.keras.Model(inputs, x)

# Copy weights
for new_l, old_l in zip(new_model.layers, old_model.layers[1:]):
    try:
        new_l.set_weights(old_l.get_weights())
    except:
        pass

# Save clean model
new_model.save("models/rice_model_clean.keras")

print("Rice model rebuilt successfully ✅")
```
